```python
import io

# define a function that reads 8 lines at at time ones and Pcess it. Until the end of the fI
def process_file(filename):
    while True:
        lines = [filename.readline() for _ in range(8)]
        
        if not any(lines):
            break
        #Pcess the lines
        lines=''.join(lines).strip().replace('\n','').replace(' ','').replace('t','u')
        #print(lines)
        #find the start codon and translate the sequence until the stop codon. Make tH recur
        seq = []
        def translate(lines, start_codon):
            
            for i in range(start_codon, len(lines)+1, 3):
                codon = lines[i:i+3]
                if codon in codon_table:
                    if codon_table[codon] == 'Stop':
                        seq.append(lines[start_codon:i])
                        lines=lines[i+3:]
                        start_codon = lines.find('aug') #start looking for the next start codon after the stop codon
                        #print(start_codon)
                        translate(lines, start_codon+3)
                        break #
        
                    
            #xxx.append(amino_acid)
            return seq
        
        start_codon = lines.find('aug')     
        seq.append(translate(lines, start_codon+3))
        
        for item in seq:
            #take codons and translate them to amino acids
            amino_acid = []
            for k in range(0, len(item), 3):
                try:
                    codon = item[k:k+3]
                    if codon in codon_table:
                        amino_acid.append(codon_table[codon])
                except:
                    pass
                
            print(''.join(amino_acid))
        print("---------------------")
            
            
            
#dictionary of codons and their corresponding amino acids
codon_table = {
    'uuu': 'F', 'uuc': 'F', 'uua': 'L', 'uug': 'L',
    'ucu': 'S', 'ucc': 'S', 'uca': 'S', 'ucg': 'S',
    'uau': 'Y', 'uac': 'Y', 'uaa': 'Stop', 'uag': 'Stop',
    'ugu': 'C', 'ugc': 'C', 'uga': 'Stop', 'ugg': 'W',
    'cuu': 'L', 'cuc': 'L', 'cua': 'L', 'cug': 'L',
    'ccu': 'P', 'ccc': 'P', 'cca': 'P', 'ccg': 'P',
    'cau': 'H', 'cac': 'H', 'caa': 'Q', 'cag': 'Q',
    'cgu': 'R', 'cgc': 'R', 'cga': 'R', 'cgg': 'R',
    'auu': 'I', 'auc': 'I', 'aua': 'I', 'aug': 'M',
    'acu': 'T', 'acc': 'T', 'aca': 'T', 'acg': 'T',
    'aau': 'N', 'aac': 'N', 'aaa': 'L', 'aag': 'L',
    'agu': 'S', 'agc': 'S', 'aga': 'R', 'agg': 'R',
    'guu': 'V', 'guc': 'V', 'gua': 'V', 'gug': 'V',
    'gcu': 'A', 'gcc': 'A', 'gca': 'A', 'gcg': 'A',
    'gau': 'D', 'gac': 'D', 'gaa': 'E', 'gag': 'E',
    'ggu': 'G', 'ggc': 'G', 'gga': 'G', 'ggg': 'G'
}
        


#open the file and call the function
filename='hexokinase.txt'
with open(filename) as f:
    process_file(f)
        
    
```
